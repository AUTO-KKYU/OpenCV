- bitwise_and : mask 영역에서 서로 공통으로 겹치는 부분 출력
- bitwise_or : mask 영역의 두 이미지를 합쳐버림


```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2

img = cv2.imread('/home/kkyu/amr_ws/opencv/data/Hawkes.jpg')

# 마스크 만들기 
mask = np.zeros_like(img)
cv2.circle(mask, (260,210), 100, (255,255,255), -1)

# 마스킹
masked = cv2.bitwise_and(img, mask)

cv2.imshow('original', img)
cv2.imshow('mask', mask)
cv2.imshow('masked', masked)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2

img = cv2.imread('/home/kkyu/amr_ws/opencv/data/Hawkes.jpg')

# 마스크 만들기 
mask = np.zeros_like(img)
cv2.rectangle(mask, (300,250), (500,50), (255,255,255), -1) # 선 두께를 -1로 하면 사각형 안이 꽉참 전체를 의미 

# 마스킹
masked = cv2.bitwise_and(img, mask)

font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(masked, text = 'hill', org = (300,250),
            fontFace=font, fontScale=2, color = (255,255,255),
            thickness=2, lineType=cv2.LINE_AA) # LINE_AA : antialiased line

cv2.imshow('original', img)
cv2.imshow('mask', mask)
cv2.imshow('masked', masked)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

img = cv2.imread('/home/kkyu/amr_ws/opencv/data/Giraffe.jpg')

giraffe_mask = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_, giraffe_mask = cv2.threshold(giraffe_mask, 127, 255, cv2.THRESH_BINARY)

masked = cv2.bitwise_and(img, img, mask=giraffe_mask)

cv2.imshow('Masked Image', masked)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
